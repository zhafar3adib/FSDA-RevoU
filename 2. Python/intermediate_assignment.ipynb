##Last Transaction per User
Find the latest transcation for every user, we will using `30 September 2021` as current date

```python
#find users who made at least one transaction
balance_last_transaction = balance[(balance['Saham_transaction_amount'] != 0) |
                              (balance['Pasar_Uang_transaction_amount'] != 0) |
                              (balance['Pendapatan_Tetap_transaction_amount'] != 0)|
                              (balance['Campuran_transaction_amount'] != 0)]

#grouping by user and find the recent transcation
balance_last_transaction = balance_last_transaction.groupby(
    'user_id', as_index=False).aggregate({'date':np.max})

balance_last_transaction.rename(columns={'date': 'recent_transaction'}, inplace=True)
balance_last_transaction
```

##Trends per Day

```python
#find users who made at least one transaction
balance_transaction_per_day = balance[(balance['Saham_transaction_amount'] != 0) |
                              (balance['Pasar_Uang_transaction_amount'] != 0) |
                              (balance['Pendapatan_Tetap_transaction_amount'] != 0)|
                              (balance['Campuran_transaction_amount'] != 0)]

#grouping by date and find amount of transcation per product
balance_transaction_per_day = balance_transaction_per_day[['date','Saham_transaction_amount',
                                                           'Pasar_Uang_transaction_amount',
                                                           'Pendapatan_Tetap_transaction_amount',
                                                           'Campuran_transaction_amount']].groupby(
                                                               'date').aggregate(np.count_nonzero)

#find how much total transcation per day
balance_transaction_per_day['how_much_transaction'] = balance_transaction_per_day.iloc[:,1:].sum(axis=1)
balance_transaction_per_day                 
```

```python
sns.set(rc={'figure.figsize':(16,9)})
sns.color_palette("tab10")
sns.lineplot(data=balance_transaction_per_day.iloc[:,:-1],
             palette='tab10',linewidth=2.5).set_title('How Much Transaction per Product per Day')
```

```python
#extract days name and grouping by them
balance_transaction_per_days = balance_transaction_per_day.reset_index()
balance_transaction_per_days['days'] = balance_transaction_per_days['date'].dt.strftime('%A')
balance_transaction_per_days = balance_transaction_per_days.groupby('days').aggregate('sum')
orderlist = ['Monday','Tuesday','Wednesday','Thursday','Friday']
balance_transaction_per_days = balance_transaction_per_days.reindex(orderlist)
balance_transaction_per_days.reset_index(inplace=True)
balance_transaction_per_days
```

```python
sns.color_palette("tab10")
sns.barplot(data=balance_transaction_per_days, x='days', y='how_much_transaction',
             palette='tab10',ci=None).set_title('Count of Transaction per Days Name')
```

```python
#define buy and sell per product by transaction amount
balance_aug['saham_buy'] = balance_aug[
    'Saham_transaction_amount'].apply(lambda x: 1 if x>0 else 0)
balance_aug['saham_sell'] = balance_aug[
    'Saham_transaction_amount'].apply(lambda x: 1 if x<0 else 0)
balance_aug['pasar_uang_buy'] = balance_aug[
    'Pasar_Uang_transaction_amount'].apply(lambda x: 1 if x>0 else 0)
balance_aug['pasar_uang_sell'] = balance_aug[
    'Pasar_Uang_transaction_amount'].apply(lambda x: 1 if x<0 else 0)
balance_aug['pendapatan_tetap_buy'] = balance_aug[
    'Pendapatan_Tetap_transaction_amount'].apply(lambda x: 1 if x>0 else 0)
balance_aug['pendapatan_tetap_sell'] = balance_aug[
    'Pendapatan_Tetap_transaction_amount'].apply(lambda x: 1 if x<0 else 0)
balance_aug['campuran_buy'] = balance_aug[
    'Campuran_transaction_amount'].apply(lambda x: 1 if x>0 else 0)
balance_aug['campuran_sell'] = balance_aug[
    'Campuran_transaction_amount'].apply(lambda x: 1 if x<0 else 0)

#find total buy and sell
balance_aug[
    'total_buy'] = balance_aug[
        'saham_buy']+balance_aug['pasar_uang_buy']+balance_aug[
            'pendapatan_tetap_buy']+balance_aug['campuran_buy']
balance_aug[
    'total_sell'] = balance_aug[
        'saham_sell']+balance_aug['pasar_uang_sell']+balance_aug[
            'pendapatan_tetap_sell']+balance_aug['campuran_sell']

#find total amount of transaction
balance_aug[
    'total_transaction'] = balance_aug[
        'Saham_transaction_amount']+balance_aug['Pasar_Uang_transaction_amount']+balance_aug[
            'Pendapatan_Tetap_transaction_amount']+balance_aug['Campuran_transaction_amount']

#find total count of transaction
balance_aug['how_much_transaction'] = balance_aug['total_buy']+balance_aug[
    'total_sell']

#create date column and fill it with maximum date for grouping
# balance_aug['date'] = balance_aug['date'].max()
balance_aug_pivot = pd.pivot_table(balance_aug,
                                     values=['Campuran_transaction_amount',
       'Pasar_Uang_transaction_amount', 'Pendapatan_Tetap_transaction_amount',
       'Saham_transaction_amount', 'saham_buy', 'saham_sell', 'pasar_uang_buy',
       'pasar_uang_sell', 'pendapatan_tetap_buy', 'pendapatan_tetap_sell',
       'campuran_buy', 'campuran_sell', 'total_buy', 'total_sell',
       'total_transaction','how_much_transaction'], index='date', aggfunc='sum')

balance_aug_pivot
```

####Transform Balance dataset

```python
#merge the dataset that already divide before
balance_endperiod = saham_clean.merge(pasar_uang_clean, how='left', on='user_id')\
.merge(pendapatan_tetap_clean, how='left', on='user_id')\
.merge(campuran_clean, how='left', on='user_id')

#remove duplicate
balance_endperiod = balance_endperiod.loc[:,~balance_endperiod.columns.duplicated()]
balance_endperiod = balance_endperiod.drop(balance_endperiod.loc[:,['date_y']], axis=1)

#assign value from Net Transaction dataframe
balance_endperiod['Saham_transaction_amount'] = net_transaction['Saham_transaction_amount']
balance_endperiod['Pasar_Uang_transaction_amount'] = net_transaction['Pasar_Uang_transaction_amount']
balance_endperiod['Pendapatan_Tetap_transaction_amount'] = net_transaction[
    'Pendapatan_Tetap_transaction_amount']
balance_endperiod['Campuran_transaction_amount'] = net_transaction[
    'Campuran_transaction_amount']

#Sum AUM, invested amount, profit/loss, count of transaction, and amount of transcation
balance_endperiod['total_AUM'] = balance_endperiod['Saham_AUM']+balance_endperiod[
    'Pasar_Uang_AUM']+balance_endperiod['Pendapatan_Tetap_AUM']+balance_endperiod['Campuran_AUM']
balance_endperiod['total_invested_amount'] = balance_endperiod['Saham_invested_amount']+balance_endperiod[
    'Pasar_Uang_invested_amount']+balance_endperiod['Pendapatan_Tetap_invested_amount']+balance_endperiod[
        'Campuran_invested_amount']
balance_endperiod['profit/loss'] = balance_endperiod['total_AUM'] - balance_endperiod['total_invested_amount']
balance_endperiod['profit/loss_percentage'] = (
    balance_endperiod['total_AUM'] - balance_endperiod['total_invested_amount'])/balance_endperiod[
        'total_invested_amount']

balance_endperiod['net_transaction'] = balance_endperiod['Saham_transaction_amount']+balance_endperiod[
    'Pasar_Uang_transaction_amount']+balance_endperiod[
        'Pendapatan_Tetap_transaction_amount']+balance_endperiod['Campuran_transaction_amount']
balance_endperiod['total_buy'] = balance_endperiod['saham_buy']+balance_endperiod[
    'pasar_uang_buy']+balance_endperiod['pendapatan_tetap_buy']+balance_endperiod['campuran_buy']
balance_endperiod['total_sell'] = balance_endperiod['saham_sell']+balance_endperiod[
    'pasar_uang_sell']+balance_endperiod['pendapatan_tetap_sell']+balance_endperiod['campuran_sell']
balance_endperiod['total_how_much_transaction'] = balance_endperiod[
    'saham_how_much_transaction']+balance_endperiod['pasar_uang_how_much_transaction']+balance_endperiod[
        'pendapatan_tetap_how_much_transaction']+balance_endperiod['campuran_how_much_transaction']

#cleaning process
balance_endperiod['profit/loss_percentage'].replace([np.inf, -np.inf], np.nan, inplace=True)
balance_endperiod.rename(columns={'date_x': 'end_of_period'}, inplace=True)
balance_endperiod = balance_endperiod.fillna(0)
balance_endperiod
```

```python
users_final.columns
```

```python
users_final_class = users_final.copy()
# users_final_class = users_final_class[(users_final_class['total_invested_amount'] !=0)]
users_final_dummy = pd.get_dummies(
    users_final_class, columns=["user_gender",'user_occupation','user_income_range','user_income_source'],
    prefix=["user_gender",'user_occupation','user_income_range','user_income_source'])
users_final_dummy.drop(users_final_dummy.iloc[:,1:38],axis=1,inplace=True)
users_final_class = users_final_class.merge(users_final_dummy)
users_final_class.drop(
    ['user_id','registration_import_datetime',"user_gender",'user_occupation','referral_code_used',
     'user_income_range','user_income_source','recent_transaction','end_of_period'],
     axis=1,inplace=True)
```

```python
numerical = ['user_age','Saham_AUM', 'Saham_invested_amount',
       'Saham_transaction_amount', 'saham_buy', 'saham_sell',
       'saham_how_much_transaction', 'Pasar_Uang_AUM',
       'Pasar_Uang_invested_amount', 'Pasar_Uang_transaction_amount',
       'pasar_uang_buy', 'pasar_uang_sell', 'pasar_uang_how_much_transaction',
       'Pendapatan_Tetap_AUM', 'Pendapatan_Tetap_invested_amount',
       'Pendapatan_Tetap_transaction_amount', 'pendapatan_tetap_buy',
       'pendapatan_tetap_sell', 'pendapatan_tetap_how_much_transaction',
       'Campuran_AUM', 'Campuran_invested_amount',
       'Campuran_transaction_amount', 'campuran_buy', 'campuran_sell',
       'campuran_how_much_transaction', 'total_AUM', 'total_invested_amount',
       'profit/loss', 'profit/loss_percentage', 'net_transaction', 'total_buy',
       'total_sell', 'total_how_much_transaction']
scaler = MinMaxScaler()
users_final_class[numerical] = scaler.fit_transform(users_final_class[numerical])
users_final_class
```

##Clustering Process

```python
distortions = []
K = range(1,11)
for k in K:
    kmeanModel = cluster.KMeans(n_clusters=k, random_state=19)
    kmeanModel.fit(users_final_class)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# Silhouette score plot
K = range(2,11)
max_K = max(K)
fig, ax = plt.subplots(int(np.ceil(max_K/2)), 2, figsize = (15,30))

for n_clusters in K:
  kmeanModel = cluster.KMeans(n_clusters, random_state=19)

  q, mod = divmod(n_clusters,2)
  sil = SilhouetteVisualizer(kmeanModel, is_fitted = False, ax = ax[q-1][mod])
  sil.fit(users_final_class)
  sil.finalize()
  print(f"For k={n_clusters}, the average silhouette score is {sil.silhouette_score_}")
```

```python
users_final_cluster = users_final.copy()
cluster_model = cluster.KMeans(n_clusters=5, random_state=19)
cluster_model.fit(users_final_class)
cluster_label = cluster_model.labels_
users_final_cluster['cluster'] = cluster_label
users_final_cluster
```
